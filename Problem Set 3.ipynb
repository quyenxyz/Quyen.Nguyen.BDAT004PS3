# Question 1

```python
# Step 1. Import the necessary libraries
import pandas as pd
# Step 2. Import the dataset from this address.
# Step 3. Assign it to a variable called users
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep="|")
print(users.head())
# Step 4. Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("mean_age_per_occupation")
print(mean_age_per_occupation)
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least 
male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / x.count())
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)

print("male_ratio_per_occupation")
print(male_ratio_per_occupation)

# Step 6. For each occupation, calculate the minimum and maximum ages
min_max_ages_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("min_max_ages_per_occupation")
print(min_max_ages_per_occupation)
# Step 7. For each combination of occupation and sex, calculate the mean age 
mean_age_per_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("mean_age_per_occupation_sex")
print(mean_age_per_occupation_sex)
# Step 8. For each occupation present the percentage of women and men
occupation_gender_counts = users.groupby(['occupation', 'gender']).size().unstack()
occupation_totals = occupation_gender_counts.sum(axis=1)
occupation_gender_percentages = occupation_gender_counts.divide(occupation_totals, axis=0) * 100
print("occupation_gender_percentages")
print(occupation_gender_percentages)
```

```python
# Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.
# Resample the data to weekly frequency, starting the week on Monday
weekly_data = data.resample('W-MON').mean()

# Slice the first 52 weeks
weekly_data = weekly_data.iloc[:52]

# Calculate statistics for each week
weekly_stats = weekly_data.agg(['min', 'max', 'mean', 'std'])

# Display the result
print(weekly_stats)
```

```python
# Step 1. Import the necessary libraries
# Step 2. Import the dataset from this address. 
# Step 3. Assign it to a variable called chipo.
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 4. See the first 10 entries
print(chipo.head(10))

# Step 5. What is the number of observations in the dataset? 
num_observations = chipo.shape[0]
print("Number of observations:", num_observations)

# Step 6. What is the number of columns in the dataset?
num_columns = chipo.shape[1]
print("Number of columns:", num_columns)

# Step 7. Print the name of all the columns. 
print("Column names:", chipo.columns.tolist())

# Step 8. How is the dataset indexed?
print("Index type:", chipo.index)

# Step 9. Which was the most-ordered item?
most_ordered_item = chipo['item_name'].mode().values[0]
print("Most-ordered item:", most_ordered_item)

# Step 10. For the most-ordered item, how many items were ordered?
most_ordered_item_orders = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("Number of items ordered:", most_ordered_item_orders)

# Step 11. What was the most ordered item in the choice_description column? 
most_ordered_choice = chipo['choice_description'].mode().values[0]
print("Most ordered item in choice_description:", most_ordered_choice)

# Step 12. How many items were orderd in total?
total_items_ordered = chipo['quantity'].sum()
print("Total items ordered:", total_items_ordered)

# Step 13.
# •	Turn the item price into a float
# •	Check the item price type
# •	Create a lambda function and change the type of item price
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))  # Removing '$' and converting to float

print("Item price type before conversion:", type(chipo['item_price'].iloc[0]))


# Check the item price type after conversion
print("Item price type after conversion:", type(chipo['item_price'].iloc[0]))


# Step 14. How much was the revenue for the period in the dataset?
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Revenue for the period:", revenue)

# Step 15. How many orders were made in the period?
num_orders = chipo['order_id'].nunique()
print("Number of orders made in the period:", num_orders)

# Step 16. What is the average revenue amount per order?
average_revenue_per_order = revenue / num_orders
print("Average revenue amount per order:", average_revenue_per_order)


# Step 17. How many different items are sold?
num_different_items = chipo['item_name'].nunique()
print("Number of different items sold:", num_different_items)
```

```python
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

# Extract relevant columns
years = data['Year']
marriages_per_capita = data['Marriages_per_1000']
divorces_per_capita = data['Divorces_per_1000']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(years, marriages_per_capita, label='Marriages per 1000', color='blue')
plt.plot(years, divorces_per_capita, label='Divorces per 1000', color='red')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number per 1000')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Filter data for years 1900, 1950, and 2000
years_of_interest = [1900, 1950, 2000]
data_filtered = data[data['Year'].isin(years_of_interest)]

# Extract marriages and divorces per capita for the selected years
marriages_per_capita = data_filtered.set_index('Year')['Marriages_per_1000']
divorces_per_capita = data_filtered.set_index('Year')['Divorces_per_1000']

# Plot the data
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = data_filtered['Year']

bar1 = plt.bar(index - bar_width/2, marriages_per_capita, bar_width, label='Marriages per 1000', color='blue')
bar2 = plt.bar(index + bar_width/2, divorces_per_capita, bar_width, label='Divorces per 1000', color='red')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number per 1000')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xticks(index, years_of_interest)

# Add legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python

# Read the data from the CSV file
data = pd.read_csv("actor_kill_counts.csv")

# Sort the data by kill count in descending order
data_sorted = data.sort_values(by='Count', ascending=True)

# Plot the data
plt.figure(figsize=(10, 6))
plt.barh(data_sorted['Actor'], data_sorted['Count'], color='skyblue')

# Add labels and title
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')

# Add labels for each bar
for index, value in enumerate(data_sorted['Count']):
    plt.text(value, index, str(value))

# Show plot
plt.tight_layout()
plt.show()
```

```python

# Read the data from the CSV file
data = pd.read_csv("roman-emperor-reigns.csv")

# Count the number of emperors who were assassinated
assassinated_count = (data['Cause_of_Death'] == 'Assassinated').sum()

# Count the number of emperors who were not assassinated
non_assassinated_count = len(data) - assassinated_count

# Create labels for the categories
labels = ['Assassinated', 'Not Assassinated']

# Create data for the pie chart
sizes = [assassinated_count, non_assassinated_count]

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Fraction of Roman Emperors Assassinated')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show plot
plt.show()
```
